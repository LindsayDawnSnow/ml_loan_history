```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Get a list of all the named colors in Matplotlib
colors = list(mcolors.CSS4_COLORS.keys())

# Plot the names of all the named colors with their corresponding color
fig, ax = plt.subplots(figsize=(10, 8))

for i, color in enumerate(colors):
    rgba = mcolors.to_rgba(color)
    ax.text(i % 10, -i * 0.6, color, color=rgba, ha='center', va='center')

ax.set_xlim(-1, 10)
ax.set_ylim(-len(colors) * 0.6, 1)
ax.axis('off')
plt.show()
```
